## 2. Methodology

In this section, each question's method and purpose will be outlined to ensure that the flow of our analyses is clear and concise. 

For ***Question 1***, We had approximately 515k hotel reviews. The reviews included both positive and negative components in separate columns. Based on the number of positive and negative word counts, we used the “review_positivity” score and associated flag “review_is_positive” to help us classify reviews as positive or negative for the purpose of this analysis. We found that the "review_is_positive" is set to 1 when the "review_positivity" exceeds 50. Then, we counted words’ frequencies from the keyword list created for both negative and positive reviews, and we removed the stopwords to eliminate the commonly occurring words in the reviews as they don't add actionable insights for business. We then created two Word Clouds using the frequencies of distinct keywords in the reviews. From the keywords detected, we captured the main reasons behind customer satisfaction and dissatisfaction. 

In ***Question 2***, there are two parts to the analysis. The only difference between the two is for the dataset and the chain_scales. In the first part we took a join with the chain scale dataset and not considered the other hotel chains, but in the second part we considered the hotels that are not part of any chain as Regular Hotels. In both parts, we firstly ran the SQL query on the postgres database. After fetching data sucessfully, we grouped the columns on the basis of hotel ID and calculated mean for review_positivity and grouping columns on basis of the chain_scale and calculating the mean. Furthermore, we created a dataframe consisting of the mean scores of chain scales, added a column for mean scores of entire dataset and finally calculated a overall score by dividing the two (all the values scaled down and in the same scale). Similarly in the second part, we created a dataframe, the addition here being 'Regular Hotels' (independent hotels). Finally, we made a bar graph for overall score v/s the hotel chain name and have kept a threshold of 100 percent for analysis of the results.

In respect to ***Question 3***, after fetching the required data from the normalized table of Review ID and their dictionary, we aggregated the review positivity rate in terms of time period. There are 3 time slices we took - Monthly, Quarterly, and Day of the week (DOW). The positivity rate was plotted as line charts to unmask any increasing or decreasing trends.

#### Q2 - Do the hotels from reputable chains have higher positivity review score? (Segmentation taken from STR chain scales dataset)

```python
# Query to create a table for analysing the positive/negative reviews that customers leave for a hotel
# also taken into consideration the chain scale of hotel for which reviews are left
# Are the hotels with more premium brand names having fewer customers' complaints? 
# (Segmentation taken from STR chain scales dataset)

queryQ2 = """
select distinct r.reviewid, r.hotelid, h.hotel, r.review_date, r.review_is_positive,
r.negative_word_counts, r.positive_word_count, r.positive_review,
r.negative_review, r.review_positivity, r.reviewer_score,
h.avg_score, s.chain_scale
from grproject.reviews_all as r
inner join grproject.hotel_dim as h
on r.hotelid = h.hotelid
inner join grproject.str_chain_list as s
on h.hotel like '%' || s.affiliation_name || '%' ;"""

cur.execute(queryQ2)
outQ2 = cur.fetchall()
```

```python
# create a df to analyze data for the Q2
reviewsQ2 = pd.DataFrame(outQ2,
                         columns=['reviewid', 'hotelid', 'hotel', 'review_date', 'review_is_positive',\
                                  'negative_word_count', 'positive_word_count', 'positive_review',\
                                  'negative_review', 'review_positivity', 'reviewer_score', 'avg_score',\
                                  'chain_scale'\
                                 ])
reviewsQ2.head()
```

```python
# Selecting all hotels via a single query

queryQ22 = """
select distinct r.reviewid, r.hotelid, h.hotel, r.review_date, r.review_is_positive,
r.negative_word_counts, r.positive_word_count, r.positive_review,
r.negative_review, r.review_positivity, r.reviewer_score,
h.avg_score
from grproject.reviews_all as r
inner join grproject.hotel_dim as h
on r.hotelid = h.hotelid;"""

cur.execute(queryQ22)
outQ22 = cur.fetchall()
```

```python
# create a df to analyze data for all hotel reviews
reviewsQ22 = pd.DataFrame(outQ22,
                         columns=['reviewid', 'hotelid', 'hotel', 'review_date', 'review_is_positive',\
                                  'negative_word_count', 'positive_word_count', 'positive_review',\
                                  'negative_review', 'review_positivity', 'reviewer_score', 'avg_score',\
                                 ])
reviewsQ22.head()
```

```python
# Calculating review score for chain hotels

chain_hotel_scores = reviewsQ2.groupby(['hotelid'])['review_positivity'].mean()
chain_hotel_scores
```

```python
# Calculating review score for only chain hotels grouped by 
chain_hotel_scores_scaled = reviewsQ2.groupby(['chain_scale'])['review_positivity'].mean()
```

```python
chain_h_scale_score = chain_hotel_scores_scaled.round(2)
chain_h_scale_score
pd.DataFrame(chain_h_scale_score)
```

```python
chain_hotel_avg_score_df = pd.DataFrame(chain_h_scale_score)
chain_hotel_avg_score_df.reset_index(inplace=True)
chain_hotel_avg_score_df.loc[:, "overall"] = chain_hotel_scores.mean()
chain_hotel_avg_score_df.loc[:, "score_compared"] = np.round(chain_hotel_avg_score_df.loc[:, "review_positivity"]/chain_hotel_scores.mean()*100,0)

chain_hotel_avg_score_df
```

```python
# Plotting the Avg. Review Positivity Score based on Hotel Grade for the chain hotels on the scale of 0 to 100

fig = plt.figure(figsize = (10, 5))

plt.bar(x=list(chain_hotel_avg_score_df.chain_scale), height = chain_hotel_avg_score_df.score_compared.values, color ='blue', width=0.4)

plt.axhline(100)

plt.xlabel("Hotel Luxury Grade", fontsize = 14)
plt.ylabel("Avg. Review Positivity Score", fontsize = 14)
plt.title("Avg. Review Positivity Score based on Hotel Grade", fontsize = 18)
plt.show()
```

```python
# Extracting data for regular hotels
# REF: https://stackoverflow.com/questions/54219055/how-to-remove-rows-from-pandas-dataframe-if-the-same-row-exists-in-another-dataf

# Merging all reviews with chain hotel reviews
df_merge = reviewsQ22.merge(reviewsQ2, on="reviewid", how="left", indicator=True)

# Cutting the Regular hotel reviews out of all hotel reviews
reg_hotel = df_merge[df_merge["_merge"] == "left_only"].drop(columns=["_merge"])
reg_hotels = reg_hotel.drop(reg_hotel.loc[:, 'hotelid_y':].columns, axis = 1)
reg_hotels.head()
```

```python
reviewsQ22_scores = reviewsQ22.groupby(['hotelid'])['review_positivity'].mean()
# Calculating the Average positivity score for each regular hotel individually 
reg_hotels_scores = reviewsQ22_scores.round(2)
reg_hotels_scores
```

```python
# Calculating the Average positivity score for each regular hotel individually 

reg_avg_score0 = reg_hotels_scores.mean()
reg_avg_score = round(reg_avg_score0, 2)
reg_avg_score
```

```python
list_reg_avg = [reg_avg_score]
df_reg_avg = pd.DataFrame(list_reg_avg, columns = ['review_positivity'])

all_hotel_avg_score_df = chain_hotel_avg_score_df.append(df_reg_avg, ignore_index = True)
all_hotel_avg_score_df.loc[5, "chain_scale"] = "Regular Hotels"
all_hotel_avg_score_df.loc[:, "overall"] = all_hotel_avg_score_df['review_positivity'].mean()
all_hotel_avg_score_df.loc[5, "score_compared"] = np.round(all_hotel_avg_score_df.loc[5, "review_positivity"]/all_hotel_avg_score_df['review_positivity'].mean()*100,0)

all_hotel_avg_score_df
```

```python
# Added Regular hotels' average positivity score to the previous plot

fig = plt.figure(figsize = (14, 5))

plt.bar(x=list(all_hotel_avg_score_df.chain_scale), height = all_hotel_avg_score_df.score_compared.values, color ='green', width=0.4)

plt.axhline(100)

plt.xlabel("Hotel Luxury Grade", fontsize = 14)
plt.ylabel("Avg. Review Positivity Score", fontsize = 14)
plt.title("Avg. Review Positivity Score based on Hotel Grade", fontsize = 18)
plt.show()
```

#### Q3 - Is there any identified seasonal trend in the positivity score?

```python
# Query to create a table for analysing the positive/negative reviews that customers leave for a hotel
# Q3

queryQ3 = """
select distinct r.reviewid, h.hotel, r.review_date, r.review_is_positive, r.positive_review,
r.negative_review, r.review_positivity, r.reviewer_score,
h.avg_score
from grproject.reviews_all as r
inner join grproject.hotel_dim as h
on r.hotelid = h.hotelid;
"""

# execute the query
cur.execute(queryQ3)
outQ3 = cur.fetchall()
```

```python
df2 = pd.DataFrame(outQ3,
                         columns=['reviewid', 'hotel', 'Review_Date', 'review_is_positive',\
                                  'positive_review',\
                                  'negative_review', 'Review_Positivity_Rate', 'reviewer_score', 'avg_score'\
                                 ])
df2.head(10)
```

```python

df2['Year'] = pd.DatetimeIndex(df2['Review_Date']).year
df2['Month'] = pd.DatetimeIndex(df2['Review_Date']).month
df2["Month_2digit"] = df2.Month.map("{:02}".format)

df2["Period"] = df2["Year"].astype(str) +'-'+ df2["Month_2digit"].astype(str)
df2.head(10)
```

```python
result_f = df2[["Period","Review_Positivity_Rate"]]
graph1 = result_f.groupby(['Period']).mean()
graph1
```

```python
plt.figure()
graph1.plot(); # peak in Jan 2016 monthly trend
```

```python
# Revised date format df

df2['Day'] = pd.DatetimeIndex(df2['Review_Date']).day

df2['quarter'] = df2['Review_Date'].dt.quarter

df2['Year_Quarter'] = df2['Year'].astype(str)+'Q'+df2['quarter'].astype(str)

df2.head(10)
```

```python
result_f2 = df2[["Year_Quarter","Review_Positivity_Rate"]]
graph2 = result_f2.groupby(['Year_Quarter']).mean()
graph2
```

```python
x = result_f2["Year_Quarter"].unique()
x
y = graph2["Review_Positivity_Rate"].to_list()
y 

plt.plot(x,y,marker="o", color='orange', alpha=0.9)
plt.xlabel("Quarter")
plt.xticks(fontsize=8)
plt.ylabel("Positivity Rate")
plt.title("Review positivity rate by Quarter")
plt.show()
# Important trend which we see the YOY decline of 2015 Q4 vs 2016 Q4 and 2016 Q1 vs 2017 Q1
```

```python
df2['day_of_week'] = df2['Review_Date'].dt.day_name()
df2.head(10)
```

```python
df2['day_of_week_in_number'] = df2['Review_Date'].dt.dayofweek

df2['DOW'] = df2['day_of_week_in_number'].astype(str)+' '+df2['day_of_week'].astype(str)

df2.head(10)
```

```python
result_f3 = df2[["DOW","Review_Positivity_Rate"]]
graph3 = result_f3.groupby(['DOW']).mean()
graph3
graph3.shape
```

```python
y = graph3["Review_Positivity_Rate"].to_list()
y
x = ['Mon', 'Tue', 'Wed' , 'Thu', 'Fri', 'Sat', 'Sun']
x 

plt.plot(x,y,marker="o", color='skyblue', alpha=0.9)
plt.xlabel("DOW")
plt.ylabel("Positivity Rate")
plt.title("Review positivity rate by Day of the Week (DOW)")
plt.show()
```
