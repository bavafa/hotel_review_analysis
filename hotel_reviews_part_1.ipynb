## 6. Table Definition
Queries defining the database schema and creation of normalized tables in our database.

```python
# connecting to the database
import os
import psycopg2
from dotenv import load_dotenv

load_dotenv()

conString = {'host':os.environ.get('DB_HOST'),
             'dbname':os.environ.get('DB_NAME'),
             'user':os.environ.get('DB_USER'),
             'password':os.environ.get('DB_PASS'),
             'port':os.environ.get('DB_PORT')}

conn = psycopg2.connect(**conString)
cur = conn.cursor()
```

```python
# Query to define schema
query = """CREATE SCHEMA IF NOT EXISTS grproject"""

# execute the query
cur.execute(query)
# commit
conn.commit()
```

```python
# Query to create a table for storing the hotel information
q1_hotel_info = """CREATE TABLE IF NOT EXISTS grproject.hotel_dim(
                        hotelid serial PRIMARY KEY,
                        hotel text NOT NULL UNIQUE,
                        address text,
                        country text,
                        state text,
                        city text,
                        lat float8,
                        lng float8,
                        ttl_num_reviews int CHECK (ttl_num_reviews > 0),
                        avg_score float4 CHECK (avg_score > 0));"""

# execute the query
cur.execute(q1_hotel_info)
# commit
conn.commit()
```

```python
# Query to create a table for storing all reviews
q2_review_main = """CREATE TABLE IF NOT EXISTS grproject.reviews_all(
                        reviewid serial PRIMARY KEY,
                        hotelid serial REFERENCES grproject.hotel_dim (hotelid),
                        room_type text,
                        room_type_level text,
                        bed_type text,
                        guest_type text,
                        trip_type text,
                        review_date timestamp NOT NULL,
                        review_is_positive bool,
                        review_positivity float8,
                        reviewer_nationality text,
                        reviewer_country text,
                        negative_review text,
                        negative_word_counts int,
                        positive_review text,
                        positive_word_count int,
                        reviewer_score int CHECK (reviewer_score > 0),
                        reviewers_total_contrib int CHECK (reviewers_total_contrib > 0));"""

# execute the query
cur.execute(q2_review_main)
# commit
conn.commit()
```

```python
# Query to create a table for storing mapping of hotels to continents
q3_hotel_list = """CREATE TABLE IF NOT EXISTS grproject.str_chain_list(
                    scale_id int PRIMARY KEY,
                    chain_scale text NOT NULL,
                    affiliation_name text REFERENCES grproject.hotel_dim (hotel),
                    parent_company text,
                    africa bool,
                    apac bool,
                    cs_america bool,
                    europe bool,
                    middle_east bool,
                    n_america_excl_us bool,
                    united_states bool);"""

# execute the query
cur.execute(q3_hotel_list)
# commit
conn.commit()
```

```python
cur.close()
conn.close()
```
